```python
stopwords = list(string.punctuation + string.digits + ' ')  
stopwords = set(stopwords)  

def preprocess(txt_token):
    clean = []  
    for token in txt_token:  
        
        if token.lower() not in stopwords:  
            token = re.sub(r'[^a-zA-Z]',"",token.lower())  
            
            if token != "":  
                clean.append(token)
            
    return clean
```

##  3.	Filter out words not present in either part of the corpus (freq==0), and compute both log-likelihood (G^2) and logratio values. Select top-15 most characteristic words for both parts of the corpus using a combination of these metrics: choose only those words with high logratio that also has high log-likelihood level (>15). How does the resulting list differ from the list obtained by “simple maths”?

```python
data_01 = data_fre.query("non_rec > 0 & rec > 0")
data_01
```

```python
def g2(a, b):
    """calculate log likelihood for two columns of frequencies"""
    c = a.sum()
    d = b.sum()
    
    E1 = c * ((a + b) / (c + d))
    E2 = d * ((a + b) / (c + d))
    
    return 2*((a*np.log(a/E1+1e-7)) + (b*np.log(b/E2+1e-7)))
```

```python
import numpy as np
part_g2 = (data_01
           .assign(g2 = lambda x: g2(x.rec, x.non_rec))
                   .sort_values('g2', ascending=False)
                   .assign(g2 = lambda x: round(x.g2, 2)))

part_g2[0:10]
```

```python
def logratio(a, b):
    ratio = (a +1) /(a.sum() + 1) / ((b + 1)/(b.sum()+1))
    return np.log(ratio)
```

```python
part_lr = (part_g2
 .sort_values('g2', ascending=False)
 .assign(g2 = lambda x: round(x.g2, 2))
 .assign(logratio = lambda x: logratio(x.rec, x.non_rec)) )
part_lr.head()
```

### top-15 most characteristic words for recommended using logratio tand log-likelihood level

```python
(part_lr
 .query('rec > 0 & non_rec > 0')
 .query('g2 > 15')
 .sort_values("logratio", ascending=False)
 .head(15))
```

### top-15 most characteristic words for non-recommended using logratio tand log-likelihood level

```python
(part_lr
 .query('rec > 0 & non_rec > 0')
 .query('g2 > 15')
 .sort_values("logratio", ascending=True)
 .head(15))
```

### top-15 most characteristic words for both recommended and non-recommended

```python
(part_lr
 .query('rec > 0 & non_rec > 0')
 .query('g2 > 15')
 .sort_values("logratio", key= lambda x: x.abs(), ascending=False)
 .head(15))
```

##  5.	Select highest rated (Rating==5) and lowest rated (Rating<3) reviews. Compile two lists of the top-100 words most closely associated with high and low review rating. You may use PMI (pointwise mutual information) or any other lexical specificity measure we discussed in this course.

```python
data_wcount
```

```python
h = data_wcount[data_wcount['Rating'] == 5]
h = h.assign(rating="high")
h
```

```python
l = data_wcount[data_wcount['Rating'] < 3]
l = l.assign(rating="low")
l
```

```python
h_l = pd.concat([h, l], ignore_index=True)
h_l
```

```python
rev = (h_l
            .assign(n=lambda x: x.groupby(['rating', 'words'])['words'].transform('count'))
            .pivot_table(values='n', columns='rating', index='words', fill_value=0)
            .reset_index())
rev
```

```python
def pmi(a, b):
    """Compute pmi of two vectors"""
    sum_a = a.sum()
    sum_b = b.sum()
    p_xy = a/sum_a
    p_x = (a + b) / (sum_a + sum_b)
    p_y = sum_a / (sum_a + sum_b)
    return (np.log((p_xy) / ((p_x * p_y))))  # (p_xy + 1)? to get rid of 0s?
```

```python
data_pmi = (rev
           .assign(pmi_high=lambda x: pmi(x.high+1, x.low+1))
           .assign(pmi_low=lambda x: pmi(x.low+1, x.high+1))
           .sort_values(by='pmi_high', ascending=False))
data_pmi.head(15) 
```
